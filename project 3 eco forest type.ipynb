### Step:3 Data Cleaning And Transformation 

```python
# Step 3: Data Cleaning & Transformation (final version)
import numpy as np

print("✅ Checking for missing values...")
print("Total missing:", Data.isna().sum().sum())  # should print 0
print("No missing values found. Proceeding...")
```

```python
# --- Handle Outliers (IQR method) ---
for col in Data.select_dtypes(include=['number']).columns:
    q1 = Data[col].quantile(0.25)
    q3 = Data[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    Data[col] = Data[col].clip(lower, upper)


# --- Fix Skewness (for distance features) ---
for col in ['Horizontal_Distance_To_Roadways', 
            'Horizontal_Distance_To_Hydrology', 
            'Horizontal_Distance_To_Fire_Points']:
    if col in Data.columns:
        Data[col] = np.log1p(Data[col])

print("✅ Data Cleaning & Transformation Completed Successfully.")
```

### Step:4 Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
import joblib

# --- 1️⃣ Create derived / interaction features (optional but useful) ---
if {'Horizontal_Distance_To_Roadways', 'Horizontal_Distance_To_Hydrology'}.issubset(Data.columns):
    Data['Road_Hydro_Ratio'] = (
        Data['Horizontal_Distance_To_Roadways'] /
        (Data['Horizontal_Distance_To_Hydrology'] + 1)   # +1 to avoid divide-by-zero
    )

if {'Hillshade_Noon', 'Hillshade_9am'}.issubset(Data.columns):
    Data['Hillshade_Diff_Noon_9am'] = (
        Data['Hillshade_Noon'] - Data['Hillshade_9am']
    )

print("Added derived features: Road_Hydro_Ratio, Hillshade_Diff_Noon_9am")

# --- 2️⃣ Encode the target variable Cover_Type ---
le = LabelEncoder()
Data['Cover_Type'] = le.fit_transform(Data['Cover_Type'])

# Save the encoder to reuse during inference (Streamlit app)
joblib.dump(le, "label_encoder.pkl")

print("✅ Label encoding completed and saved as label_encoder.pkl")

# --- 3️⃣ Check final shape ---
print("Data shape after feature engineering:", Data.shape)
print("Columns:", Data.columns.tolist()[:10], "...")  # preview first 10
```

### Step:6 Class Imbalance Handling 

```python
from imblearn.over_sampling import RandomOverSampler

# Separate features and target
X = Data.drop('Cover_Type', axis=1)
y = Data['Cover_Type']

# Check imbalance before
print("Before balancing:\n", y.value_counts())

# Apply RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)

# Check after balancing
print("\nAfter balancing:\n", pd.Series(y_res).value_counts())
```

### Step:7 Feature Selection 

```python
# (Most Important Columns) - Optimized

from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# Separate features and target
y = Data['Cover_Type']
X = Data.drop('Cover_Type', axis=1)

# 🔹 Use a smaller sample of the data for speed
X_sample = X.sample(20000, random_state=42)
y_sample = y.loc[X_sample.index]

# 🔹 Train a smaller Random Forest using all CPU cores
rf = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
rf.fit(X_sample, y_sample)

# Get feature importance scores
importances = pd.Series(rf.feature_importances_, index=X.columns)

# Sort and view top 15 features
top_features = importances.sort_values(ascending=False).head(15)
print("Top 15 Most Important Columns:")
print(top_features)

# Plot top features
plt.figure(figsize=(8,5))
top_features.sort_values().plot(kind='barh', color='skyblue')
plt.title("Top 15 Important Features (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Feature Name")
plt.tight_layout()
plt.show()
```

### Step:8 Model Building

```python
# step 8.1 – Train different ML models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42
)

# Scale only for Logistic Regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define models
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    "Logistic Regression": LogisticRegression(max_iter=2000, n_jobs=-1),
    "KNN": KNeighborsClassifier(),
    "XGBoost": xgb.XGBClassifier(eval_metric="mlogloss", random_state=42)
}
```

```python
# Step 8.2 – Evaluate models
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

results = {}

for name, model in models.items():
    print(f"\n🔹 Training {name} ...")
    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        preds = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, preds))

    # Confusion Matrix
    cm = confusion_matrix(y_test, preds)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
    plt.title(f"{name} – Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

print("\n✅ Model accuracy comparison:")
for k, v in results.items():
    print(f"{k}: {v:.4f}")
```

```python
# Step 8.3 – Hyperparameter tuning for Random Forest
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100],
    'max_depth': [10,None],
    'min_samples_split': [2],
    'min_samples_leaf': [1]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42, n_jobs=-1),
    param_grid=param_grid,
    cv=3,
    scoring='accuracy',
    verbose=2,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
print("✅ Best parameters:", grid_search.best_params_)

best_rf = grid_search.best_estimator_

# Evaluate tuned model
preds = best_rf.predict(X_test)
print("\n🎯 Tuned RF Accuracy:", accuracy_score(y_test, preds))
print(classification_report(y_test, preds))

# Confusion matrix for tuned model
cm = confusion_matrix(y_test, preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
plt.title("Tuned Random Forest – Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
